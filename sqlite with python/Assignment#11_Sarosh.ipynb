```python
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
def sql_query(q):
    return pd.read_sql_query(q, conn)
```

### ***7. Show All Employess With A Total Pay Between 50,000 and 75,000***

```python
q = '''select BasePay , OvertimePay, OtherPay, Benefits, TotalPay, TotalPayBenefits  from salaries where BasePay is null

or  OvertimePay is null or  OtherPay is null or  Benefits is null or TotalPay is null or  TotalPayBenefits is null
'''
sql_query(q)  # no null value in NUMERIC data
```

### ***8. Show All Employees With A BasePay less than 50,000 Or A TotalPay Greater Than 100,000***

```python
q = '''select EmployeeName ,BasePay, TotalPay  from salaries where 
         BasePay is  null
        or
        BasePay = 'Not Provided'
'''
sql_query(q)  ## no Null and not provided value
```

### ***9. Show All Employees With A TotalPay Benefits Value Between 125,000 and 150,000 And A Job Title Containing Word 'Director'***
