# Logical Operators in Pandas:
    * & (and)
    * | (or)
    * ~ (not)

```python
df['State'] == 'Florida'
```

```python
df[df['State'] == 'Florida']
```

```python
df['Category'].unique()
```

```python
df[df['Category'] == 'Technology'] 
```

```python
df[(df['State'] == 'Florida') & (df['Category'] == 'Technology')]
```

```python
df[(df['State'] == 'Florida') | (df['State'] == 'California')]
```

```python
(df['State'] == 'Florida') ~(df['Category'] == 'Technology')
```

```python
(df['State'] == 'Florida') & ~(df['Category'] == 'Technology')
```

```python
df[(df['State'] == 'Florida') & ~(df['Category'] == 'Technology')]
```

```python
df[(df['State'] == 'Florida') | ~(df['Category'] == 'Technology')]
```
