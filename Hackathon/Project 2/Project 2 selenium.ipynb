```python
# Install necessary libraries and drivers
!pip install selenium webdriver-manager > /dev/null 2>&1
!apt-get update > /dev/null 2>&1
!apt-get install -y chromium-chromedriver > /dev/null 2>&1

# Import required modules
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# Configure Chrome options for headless browsing
browser_options = Options()
browser_options.add_argument('--headless')  # Enable headless mode
browser_options.add_argument('--no-sandbox')  # Disable sandbox for better compatibility
browser_options.add_argument('--disable-dev-shm-usage')  # Resolve resource limitations
browser_options.add_argument('--disable-gpu')  # Optional for rendering

# Path to the ChromeDriver installed in the environment
driver_path = '/usr/bin/chromedriver'

# Initialize the WebDriver
browser = webdriver.Chrome(options=browser_options)

# Automate Flipkart search
try:
    # Navigate to Flipkart
    browser.get("https://www.flipkart.com/")
    print("Website Title:", browser.title)

    # Handle the initial popup (if displayed)
    try:
        dismiss_button = browser.find_element(By.XPATH, "//button[contains(text(),'✕')]")
        dismiss_button.click()
    except Exception as error:
        print("Popup not found or already closed:", error)

    # Search for "one plus headphones"
    search_input = browser.find_element(By.XPATH, "//input[@type='text']")
    search_input.send_keys("one plus headphones")
    search_input.submit()

    # Wait for results to load and find elements
    browser.implicitly_wait(10)
    product_elements = browser.find_elements(By.CLASS_NAME, "wjcEIp")

    # Extract URLs and titles for all products
    product_details = [{"link": item.get_attribute("href"), "name": item.text} for item in product_elements]

    # Locate the product image using a specific CSS selector
    product_image = browser.find_element(By.CSS_SELECTOR, ".VJA3rP > div > div > div._4WELSP > img.DByuf4")
    image_url = product_image.get_attribute("src")

    # Locate the price of the product using another CSS selector
    product_price = browser.find_element(By.CSS_SELECTOR, "a>.hl05eU>div.Nx9bqj").text

    # Print details for each product
    for idx, product in enumerate(product_details, start=1):
        print(f"Product {idx} Link: {product['link']}")
        print(f"Product {idx} Name: {product['name']}")
        print(f"Price: {product_price}")  # Price will be the same for all items
        print(f"Image {idx} URL: {image_url}")  # Single image for all items

    # Display the current page URL
    print("Current Page URL:", browser.current_url)

finally:
    # Close the browser instance
    browser.quit()
```
