## ***5.Check Null Values And Also Create A Heatmap In The Dataset***


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (assuming it's already loaded as 'data')
# data = pd.read_csv('/path/to/adult.csv')

# 1. Check for null values in each column
null_values = df.isnull().sum()
print("Null values in each column:")
print(null_values)

# 2. Create a heatmap to visualize the missing data
plt.figure(figsize=(10, 6))  # Set figure size
sns.heatmap(df.isnull(), cmap="viridis", cbar=False)
plt.title("Heatmap of Missing Values")
plt.show()
```

## ***12. What Is The Distribution of Age Column Show In Graph?***


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the size of the plot
plt.figure(figsize=(10, 6))

# Plotting the distribution of the 'age' column
sns.histplot(df['age'], bins=30, kde=True, color='blue')

# Adding titles and labels
plt.title('Distribution of Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show the plot
plt.show()
```

## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
age_between_17_and_48 = df[df['age'].between(17, 48)]


total_between = age_between_17_and_48.shape[0]

print(f"Total number of persons having age between 17 to 48 using between is: {total_between}")


age_between_17_and_48_no_between = df[(df['age'] >= 17) & (df['age'] <= 48)]


total_without_between = age_between_17_and_48_no_between.shape[0]

print(f"Total number of persons having age between 17 to 48 without  between is : {total_without_between}")
```

## ***14. What is The Distribution of Workclass Column***

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


workclass_distribution = df['workclass'].value_counts()

print("Distribution of Workclass column:")
print(workclass_distribution)

# Plot the distribution as a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=workclass_distribution.index, y=workclass_distribution.values, palette='viridis')
plt.title("Distribution of Workclass Column")
plt.xlabel("Workclass")
plt.ylabel("Frequency")

plt.show()
```

## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python

df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})


print("Updated Income Column:")
print(df['income'].value_counts())
```

## ***18. Which Workclass getting The Highest Salary***

```python

highest_salary_takers = df.groupby('workclass')['income'].mean().idxmax()

print(f"The workclass getting the highest salary is: {highest_salary_takers}")
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
gender_income_distribution = df.groupby('gender')['income'].mean()

print(gender_income_distribution)

better_chance_gender = gender_income_distribution.idxmax()

print(f"The gender with a better chance of earning >50K is: {better_chance_gender}")
```
