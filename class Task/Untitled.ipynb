```python
# import important libraries
import pandas as pd
import sqlite3 as sql
```

```python
q = '''SELECT c.customer_id,
       c.country,
       iv.total AS top_buyer,
       e.first_name,
       e.last_name
FROM customer AS c
INNER JOIN employee AS e ON e.employee_id = c.support_rep_id
INNER JOIN invoice AS iv ON c.customer_id = iv.customer_id
WHERE (iv.total) = (
    SELECT MAX(iv_sub.total)
    FROM invoice AS iv_sub
    WHERE iv_sub.customer_id = c.customer_id
)
GROUP BY c.customer_id, c.country, iv.total, e.first_name, e.last_name
ORDER BY c.country, iv.total DESC;



'''
sql_query(q) 
```

```python
q = '''select c.customer_id,  c.country , CONCAT(c.first_name, ' ', c.last_name) AS full_name , sum (iv.total) as Top_buyer from employee as e
inner join customer as c 
on e.employee_id =c.support_rep_id
inner join invoice as iv
on c.customer_id=iv.customer_id
group by c.country
having  iv.total in (select max (iv.total) from invoice as iv 
inner join   customer as c
on c.customer_id=iv.customer_id 
group by iv.customer_id  )

'''
sql_query(q)  
```
