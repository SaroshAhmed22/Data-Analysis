```python
# import important libraries
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("chin.db")
conn
```

```python
def sql_query(q):
    return pd.read_sql_query(q, conn)
```

```python
q = '''select * from sqlite_master where type='table'
'''
sql_query(q) 
```

```python
q = '''select * from invoice limit 1
'''
sql_query(q)  # no null value in NUMERIC data
```

```python
q = '''select c.customer_id,  c.country , iv.total as top_buyer from employee as e
inner join customer as c 
on e.employee_id =c.support_rep_id
inner join invoice as iv
on c.customer_id=iv.customer_id
group by c.country
having  iv.total in (select max (iv.total) from invoice as iv 
inner join   customer as c
on c.customer_id=iv.customer_id 
group by c.customer_id  )

'''
sql_query(q)  
```

```python
q = '''
select max (iv.total) from invoice as iv 
inner join   customer as c
on c.customer_id=iv.customer_id 
group by c.customer_id 

'''
sql_query(q)  
```
